# Notebook to compute the Apriori data selection

This is the first notebook to run to perform the apriori data selection. 
Other than the apriori data selection, this notebook creates the tabular dataset composed by all signals after the apriori signal selection. 

NB: If a tabular dataset is already avaiable this step is not required. 

Input: 
 - <cycle_order.csv> the file with the list of cycles and their labels
 - <fpath> the path where cycle data and cycle_order.csv are stored

Output: 
- <selected/All.pkl> the list of signals after the Apriori selection
- <dataset/All.pkl> the dataset for the Supervised Signal Selection

```python
import pandas as pd
import pathlib
import pickle
import sys
import os

%matplotlib inline


ParentPath = str(pathlib.Path().absolute().parent)
sys.path.insert(1, ParentPath+'/1-SignalSelection/functions')

import APrioriSelection 


#path of the cycles 
fpath = ParentPath+'/data/'

file_label = pd.read_csv(fpath+'cycle_order.csv', sep=",")


# APriori Data Selection                                          
# Inside of functions/APrioriSelection.py it is possible to specify which signals should be discarded according to domain-knowledge
# Parameters: 
    # file_label['cycleName']: the list of cycles ordered by acquisition time
    # fpath: Path with the row cycles 
    # o2_considerated: Consider the o2 signal or not
    # strong_corr_considerated: Remove strongly correlated signals
#Output: Remaining Signals
selection = APrioriSelection.APriori(list(file_label['cycleName']), fpath, o2_considerated=False, strong_corr_considerated = False)
signalSelected = selection.select()

if not os.path.exists('selected'):
    os.makedirs('selected')    

pickle.dump(signalSelected,open("selected/All.pkl","wb"))
```

```python
sys.path.insert(1, ParentPath+'/classes/public/')
from makerDatasetSpecialized import MakerDatasetSpec


if not os.path.exists('dataset'):
    os.makedirs('dataset')    

signalsSelected = pickle.load(open("selected/All.pkl","rb"))

DeltaT = 60
maker = MakerDatasetSpec(file_label, fpath, minuteWindow = DeltaT)
Xnp, Ynp, Id = maker.makeDataset(signals=signalsSelected)
df = pd.DataFrame.from_records(Xnp)
df.columns = maker.features
df.insert(loc=0, column='ExpID', value=Id)
df['Label']=Ynp
pickle.dump(df,open("dataset/All.pkl","wb")) 
```
