```python
import sys
import os
from collections import OrderedDict
import numpy as np
import json
import random

def  createForestGrid(out_dir = './'):
    n_jobs = -1
    max_depth = [30, 50, None]
    min_samples_split =  [2, 20]
    bootstrap = [True,False]
    
    n_estimators = [10,50,100,500,1000,1500]
    min_impurity_decrease = [0.0,0.005,0.01,0.02]
    min_samples_leaf = [1,5,10,20]
    random_state = [42]
    criteria = ['gini']
    max_features = ['auto','log2',None]

    conf_list = []
    for max_depthi in max_depth:
        for min_samples_spliti in min_samples_split:
            for bootstrapi in bootstrap:
                for min_imp in min_impurity_decrease:
                    for leaf in min_samples_leaf:
                        for n_est in n_estimators:
                            for rs in random_state:
                                for criterion in criteria:
                                    for max_featuresi in max_features:
                                        conf = dict(n_jobs=n_jobs, max_depth=max_depthi,\
                                                    min_samples_split=min_samples_spliti, bootstrap=bootstrapi,
                                                    min_impurity_decrease=min_imp, min_samples_leaf=leaf,\
                                                    n_estimators=n_est, criterion=criterion,\
                                                    random_state=rs, max_features= max_featuresi)
                                        conf_list.append(conf)

    print('RF Total Conf:', len(conf_list))
    with open(os.path.join(out_dir, 'grid_forest.json'), 'w') as fp:
        json.dump(conf_list, fp)
    return

def  createTreeGrid(out_dir = './'):

    max_depth = [30, 50, None]
    min_samples_split =  [2, 20]
    
    min_impurity_decrease = [0.0,0.005,0.01,0.02]
    min_samples_leaf = [1,5,10,15,20,25,30]
    random_state = [42]
    criteria = ['entropy','gini']
    max_features = ['auto','log2',None,0.5]

    conf_list = []
    for max_depthi in max_depth:
        for min_samples_spliti in min_samples_split:
            for min_imp in min_impurity_decrease:
                for leaf in min_samples_leaf:
                    for rs in random_state:
                        for criterion in criteria:
                            for max_featuresi in max_features:
                                conf = dict(max_depth=max_depthi,\
                                            min_samples_split=min_samples_spliti,\
                                            min_impurity_decrease=min_imp, min_samples_leaf=leaf,\
                                            criterion=criterion,\
                                            random_state=rs, max_features= max_featuresi)
                                conf_list.append(conf)

    print('Tree Total Conf:', len(conf_list))
    with open(os.path.join(out_dir, 'grid_tree.json'), 'w') as fp:
        json.dump(conf_list, fp)
    return

def createSVMGrid(out_dir = './'):
    #gamma va da 0.001 a  1000 diviso in 20 parti logaritmicamente
    gamma = list(np.logspace(-3, 3, 100, True, 10))
    #C va da 0.001 a 1000 diviso in 20 parti logarimicamente
    C = list(np.logspace(-3,3,100, True, 10))
    kernel = ['rbf','poly','linear']
    random_states = [2]
    #random.seed(0)
    #random_states = [random.randint(0, 2 ** 32 - 1) for i in range(0, 100)]
    class_weight = None

    conf_list = []
    for k in kernel:
        for c in C:
            for g in gamma:
                for random_state in random_states:
                    conf = dict(C=c, gamma=g, kernel=k, random_state=random_state, class_weight=class_weight)
                    conf_list.append(conf)

    print('SVM Total Conf:', len(conf_list))
    with open(os.path.join(out_dir, 'grid_svm.json'), 'w') as fp:
        json.dump(conf_list, fp)
    return

def main():

    for model in ["svm"]:    
        if model == 'tree':
            createTreeGrid()
        elif model == 'forest':
            createForestGrid()
        elif model == 'svm':
            createSVMGrid()

        else:
            print('Not supported classification model')
            sys.exit(-1)
    return
main()
```
